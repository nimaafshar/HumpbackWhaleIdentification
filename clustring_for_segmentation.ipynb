```python
import os
train_files = os.listdir('./train')
```

```python
import random
RANDOM_COUNT = 20
random_sample = random.sample(train_files,RANDOM_COUNT)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
BASE_FILEPATH = './train/'
import cv2 as cv
img = cv.imread(BASE_FILEPATH+random_sample[1])
```

```python
plt.imshow(img)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
```

```python
new_img = img.reshape(-1,3)
# kmeans.fit(img)
```

```python
kmeans.fit(new_img)
```

```python
newpic = kmeans.cluster_centers_[kmeans.labels_]
```

```python
v = img.shape
ix = (newpic.reshape(v[0],v[1],v[2]))/256
plt.imshow(ix)
```

```python
kernel = np.ones((7,7),np.float32)/49
dst = cv.filter2D(ix,-1,kernel)
```

```python
plt.imshow(dst)
```

```python
kmeans.fit((ix*256).reshape(-1,3))
```

```python
newpic = kmeans.cluster_centers_[kmeans.labels_]
v = img.shape
plt.imshow((newpic.reshape(v[0],v[1],v[2]))/256)
```

```python
random_sample[0]
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
for i in range(RANDOM_COUNT):
    img = cv.imread(BASE_FILEPATH+random_sample[i])
    new_img = img.reshape(-1,3)
    kmeans.fit(new_img)
    newpic = kmeans.cluster_centers_[kmeans.labels_]
    v = img.shape
    ix = (newpic.reshape(v[0],v[1],v[2]))/256
    plt.imsave('./kmeans/'+random_sample[i],ix)
    print(f'item {i} done')
```
