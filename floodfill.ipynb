```python
import cv2 as cv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
import os
covers = os.listdir('./mask')
```

```python
covers = pd.Series(covers)
```

```python
covers = covers.apply(lambda x:x[:-4])
```

```python
#reading 
def real_image_add(name):
    return './train/'+name+'.jpg'
def cover_image_add(name):
    return './mask/'+name+'.png'
def save_image_add(name):
    return './floodfilled/'+name+'.png'

img = cv.imread(cover_image_add(covers[3]),cv.IMREAD_GRAYSCALE)

h,w = img.shape
mask = np.zeros((h+2, w+2), np.uint8)
cv.floodFill(img, mask, (0,0), 255);
saveit = cv.bitwise_not(img)
plt.imsave(save_image_add(covers[3]),saveit,cmap=cm.gray)
```

```python
real_img  = cv.imread(real_image_add(covers[3]))
```

```python
mask = cv.bitwise_not(img)
plt.imshow(cv.bitwise_and(real_img,real_img,mask=mask))
```

```python
def floodfill_it(name):
    img = cv.imread(cover_image_add(name),cv.IMREAD_GRAYSCALE)
    h,w = img.shape
    mask = np.zeros((h+2, w+2), np.uint8)
    cv.floodFill(img, mask, (0,0), 255);
    saveit = cv.bitwise_not(img)
    plt.imsave(save_image_add(name),saveit,cmap=cm.gray)
```

```python
for i in covers:
    floodfill_it(i)
```
