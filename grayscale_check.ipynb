```python
import numpy as np
RANDOM_COUNT = 20
def is_grayscale(img):
    height , width , channels = img.shape
    h  = np.random.randint(height,size=20)
    w = np.random.randint(width,size=20)
    for i in range(RANDOM_COUNT):
        r , g, b = img[h[i],w[i]]
        if abs(r-g)>5 or abs(g-b)>5 :
            return False
    return True
    
```

```python
BASIC_PATH = './train/'
def is_grayscale_by_filename(filename):
    path = BASIC_PATH+filename
    image = cv.imread(path)
    return is_grayscale(image)
```
