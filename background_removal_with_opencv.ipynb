```python
import cv2 as cv
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
l = os.listdir('./train')
```

```python
sample_img = cv.imread('./train/'+l[1])
```

```python
plt.imshow(sample_img)
```

```python
mask = np.zeros(sample_img.shape[:2],np.uint8)
```

```python
bgdModel = np.zeros((1,65),np.float64)
fgdModel = np.zeros((1,65),np.float64)
h , w = sample_img.shape[:2]
rect = (90,0,250,760)
cv.grabCut(sample_img,mask,rect,bgdModel,fgdModel,5,cv.GC_INIT_WITH_RECT)
mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
sample_img = sample_img*mask2[:,:,np.newaxis]
plt.imshow(sample_img)
```
